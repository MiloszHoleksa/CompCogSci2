```python
# Function which creates a list of pairs where 'modality' == 0 is followed by 'modality' == 1
# 0 - audio-visual stimuli
# 1 - visual stimuli
# so each pair corresponds to one observation in which audio-visual stimuli is followed by visual one

def modality_pairs(df):
    pairs = []
    i = 0
    while i < len(df) - 1:
        if df.loc[i, 'modality'] == 0:

            # Skipping the first 0 if followed by another 0
            if i < len(df) - 2 and df.loc[i + 1, 'modality'] == 0:
                i += 1

            # Going to the next row if the participant is changing between consecutive rows
            elif df.loc[i, 'SubId'] != df.loc[i + 1, 'SubId']:
                i += 1

            # If we have a pair (0, 1)
            elif df.loc[i + 1, 'modality'] == 1:
                pairs.append([df.iloc[i], df.iloc[i + 1]])
                i += 2
                
            # Going to the next row if the next modality value is 0 for the same participant
            else:
                i += 1

        # Going to the next row if participant is changing
        else:
            if df.loc[i, 'SubId'] != df.loc[i + 1, 'SubId']:
                i += 1
            
        # Moving to the next row
        i += 1

    return pairs

pairs = modality_pairs(df)
pairs
```

```python
pairs_combined = []

for pair in pairs:
    pair_dict = {}
    for idx, row in enumerate(pair):
        for col_name, value in row.items():
            new_col_name = f'{col_name}_A' if idx != 0 else col_name
            pair_dict[new_col_name] = value
    pairs_combined.append(pair_dict)

# Creating a new dataframe from the list of dictionaries
df_combined = pd.DataFrame(pairs_combined)

df_combined
```

```python
import pymc as pm


# Model 1: Ventriloquist Effect
with pm.Model() as ventriloquist_model:
    # Priors
    alpha = pm.Normal('alpha', mu=0, sigma=10)
    beta_resp = pm.Normal('beta_resp', mu=0, sigma=1)
    #beta_delta = pm.Normal('beta_delta', mu=0, sigma=1)

    sigma = pm.HalfCauchy('sigma', beta=5)
    
    # Multimodal prior receptivity
    #theta = pm.Normal('theta', mu=0, sigma=1)

    # Likelihood
    mu = alpha + beta_resp * df_cleaned['Relative_response'] #+ beta_delta * df_cleaned['DeltaAV']
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed= df_cleaned['Relative_responses_A'])
    
    # Sampling
    trace_ventriloquist = pm.sample(return_inferencedata=True, idata_kwargs={"log_likelihood": True})
```

```python
# Model 2: Multimodal prior


with pm.Model() as multimodal_model:
    # Priors
    alpha = pm.Normal('alpha', mu=0, sigma=10)
    beta_resp = pm.Normal('beta_resp', mu=0, sigma=1)
    beta_prev = pm.Normal('beta_prev', mu=0, sigma=1)

    sigma = pm.HalfCauchy('sigma', beta=5)
    
    # Multimodal prior receptivity
    #theta = pm.Normal('theta', mu=0, sigma=1, shape=n_participants)

    # Likelihood
    mu = alpha + beta_resp * df_cleaned['Relative_response'] + beta_prev * df_cleaned['Previous_Relative_response']
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed= df_cleaned['Relative_responses_A'])
    
    # Sampling
    trace_multimodal = pm.sample(return_inferencedata=True, idata_kwargs={"log_likelihood": True})
```
